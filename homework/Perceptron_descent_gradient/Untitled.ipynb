# Perceptron

```python
perceptron = Algo(iris_X, iris_y, algo=Perceptron, yielding_pas=10)
perceptron.do_all(nb_iterations=MAX_PLOT_X_PERCEPTRON,name="perceptron iris", max_x=MAX_PLOT_X_PERCEPTRON)
```

```python
perceptron = Algo(mushroom_X, mushroom_y, algo=Perceptron, yielding_pas=10)
perceptron.do_all(nb_iterations=MAX_PLOT_X_PERCEPTRON,name="perceptron mushroom", max_x=MAX_PLOT_X_PERCEPTRON)
```

```python
perceptron = Algo(spambase_X, spambase_y, algo=Perceptron, yielding_pas=10)
perceptron.do_all(nb_iterations=MAX_PLOT_X_PERCEPTRON,name="perceptron spambase", max_x=MAX_PLOT_X_PERCEPTRON)
```

# Descent gradient
