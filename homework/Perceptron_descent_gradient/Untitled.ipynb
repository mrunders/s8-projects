```python
from main import *
from PlotBuilder import *
```

```python
MAX_PLOT_X = 4500

iris_X, iris_y = Data.load_iris_data()
mushroom_X, mushroom_y = Data.load_mushroom_data()
spambase_X, spambase_y = Data.load_spambase_data()
```

```python
Plot_builder.y_repartition(iris_y, title="repartition of Iris (-1=others, 1=Iris_versicolor)")
```

```python
Plot_builder.y_repartition(mushroom_y, title="repartition of Mushrooms (-1=poisonous, 1=edible)")
```

```python
Plot_builder.y_repartition(spambase_y, title="repartition of Spambase (-1=spam, 1=not a spam)")
```

```python
perceptron = Algo(iris_X, iris_y, algo=Perceptron, yielding_pas=10, shuffle_data=True)
perceptron.run(nb_iterations=MAX_PLOT_X)
perceptron_plot = perceptron.get_plot()
perceptron_plot.draw_plot(name="perceptron iris", max_x=MAX_PLOT_X)
perceptron_plot.show_plot()
```

```python
perceptron = Algo(mushroom_X, mushroom_y, algo=Perceptron, yielding_pas=10, shuffle_data=True)
perceptron.run(nb_iterations=MAX_PLOT_X)
perceptron_plot = perceptron.get_plot()
perceptron_plot.draw_plot(name="perceptron mushroom", max_x=MAX_PLOT_X)
perceptron_plot.show_plot()
```

```python
perceptron = Algo(spambase_X, spambase_y, algo=Perceptron, yielding_pas=10, shuffle_data=True)
perceptron.run(nb_iterations=MAX_PLOT_X)
perceptron_plot = perceptron.get_plot()
perceptron_plot.draw_plot(name="perceptron spambase", max_x=MAX_PLOT_X)
perceptron_plot.show_plot()
```
